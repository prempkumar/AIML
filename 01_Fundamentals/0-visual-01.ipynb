```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline  
# makes the charts render in the page and not just create those objects

import seaborn as sns
sns.set(color_codes=True)  
# rich color for background and graph
```

```python
df = pd.read_csv('Automobile.csv')
df.head()
# df.shape
# df.dtypes
```

```python
sns.distplot(df.highway_mpg);
```

```python
sns.distplot(df.highway_mpg, kde=False, rug=True);
```

```python
sns.jointplot(df.horsepower, df.engine_size);
# default is 'scatter' plot
```

```python
sns.jointplot(df.horsepower, df.engine_size, kind='reg');
```

```python
sns.jointplot(df.horsepower, df.engine_size, kind='hex');
# scatter" | "reg" | "resid" | "kde" | "hex" 
```

```python
sns.jointplot(df.horsepower, df.engine_size, kind='kde');
```

```python
sns.pairplot(df[['normalized_losses', 'engine_size', 'horsepower']]);
```

### Categorical variables

```python
sns.stripplot(df.fuel_type, df.horsepower);
#jitter = True by default
```

```python
sns.swarmplot(df.fuel_type, df.horsepower);
```

```python
sns.boxplot(df.fuel_type, df.horsepower);
```

```python
sns.boxplot(df.fuel_type, df.horsepower, hue=df.drive_wheels);
```

```python
sns.barplot(df.fuel_type, df.horsepower, hue=df.drive_wheels);
```

```python
sns.countplot(df.body_style, hue=df.drive_wheels);
```

```python
sns.pointplot(df.body_style, df.horsepower, hue=df.drive_wheels, join=False);
```

```python
sns.catplot(x='body_style', y='horsepower', col='drive_wheels', hue='number_of_doors', data=df,  kind='box');
```

```python
sns.catplot(x='body_style', y='horsepower', col='drive_wheels', hue='number_of_doors', data=df,  kind='violin');
# kind=violin is the kde chart
```

```python
sns.catplot(x='body_style', col='drive_wheels', hue='number_of_doors', data=df,  kind='count');
# kind can be one of "point", "bar", "strip", "swarm", "box", "violin", "boxen" or "count"; in the event of count
# the y value becomes count and hence y value shouldn't be provided as argument
```

```python
sns.catplot(x='body_style', y='horsepower', col='drive_wheels', hue='number_of_doors', data=df,  kind='point', join=False);
# kind can be one of "point", "bar", "strip", "swarm", "box", "violin", "boxen" or "count"; in the event of count
# the y value becomes count and hence y value shouldn't be provided as argument
```

```python
sns.lmplot(x='horsepower', y='engine_size', data=df);
```

```python
sns.lmplot(x='horsepower', y='engine_size', data=df, hue='fuel_type');
```
