```python
dfHDI = df[['country','year','country-year','HDI','GDP','percapGDP']].copy()
dfHDI = dfHDI.drop_duplicates()
dfHDI['corrHDI'] = dfHDI.HDI
dfHDI['interpHDI'] = dfHDI.HDI
```

```python
dfHDI.corr()
```

```python
dfHDI.interpHDI.interpolate(method='linear',limit_direction='forward',inplace=True)
```

```python
df = pd.merge(df,dfHDI[['country-year','interpHDI']],how="left",on="country-year")
```
