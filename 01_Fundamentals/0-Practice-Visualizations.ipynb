**size**: a numeric vector, number of people in a table/Party

### Q9] At what time of which day does the restaurant get higher amount of bill (consider for male and female separately)

use lmplot

```python
# If linear regression plot is to be used,String datatype cannot be used and hence converting Days to numeric values
def s2n(str) :
    week = {
        "Mon":1,
        "Tue":2,
        "Wed":3,
        "Thur":4,
        "Fri":5,
        "Sat":6,
        "Sun":7
    }
    return week.get(str)
df['day_num'] = df.day.apply(s2n)

sns.lmplot(x='day_num', y='total_bill', data=df, hue='time');

# However alternate plots can be used to better visualize this plot - shown below
```

```python
sns.boxplot(df.day, df.total_bill, hue=df.time);
```

```python
sns.stripplot(df.day, df.total_bill, hue=df.time);
```
